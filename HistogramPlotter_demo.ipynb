# Histogram plots with shared bins

```python
import numpy as np

import matplotlib.pyplot as plt

from hyperpyper.utils import HistogramPlotter
```

## Generating sample data

```python

data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)
data3 = np.random.normal(loc=2.5, scale=1, size=2000)
```

## Basic Plotting

```python
fig = HistogramPlotter(data1).plot()
```

```python
fig = HistogramPlotter(data1, 'data1').plot()
```

```python
fig = HistogramPlotter([data1, data2],
                      ["data1", "data2"],
                      ).plot()
```

### Number of bins

```python
fig = HistogramPlotter([data1, data2],
                      ["data1", "data2"],
                      bins=50,
                      ).plot()
```

## Color Modifications

### Use a specific colormap

```python
fig = HistogramPlotter([data1, data2],
                      ["data1", "data2"],
                      colormap='tab10',
                      ).plot()
```

```python
fig = HistogramPlotter([data1, data2],
                      ["data1", "data2"],
                      colormap=plt.get_cmap('tab10'),
                      ).plot()

                      
```

### Explicitly pass colors for each plot

```python
fig = HistogramPlotter([data1, data2],
                      ["data1", "data2"],
                      colors=['orange', 'green'],
                      ).plot()
```

### Change the colors globally

```python
# Set the custom color cycle as the default
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['red', 'blue', 'green', 'orange', 'purple'])

fig = HistogramPlotter([data1, data2, data3],
                      ["data1", "data2", "data3"],
                      density=True,
                      ).plot()
```

```python
# Reset plt.rcParams['axes.prop_cycle'] to its default value again
plt.rcParams['axes.prop_cycle'] = plt.rcParamsDefault['axes.prop_cycle']
```

## Compare data with different size

```python
fig = HistogramPlotter([data1, data2, data3],
                      ["data1", "data2", "data3"],
                      ).plot()
```

```python
fig = HistogramPlotter([data1, data2, data3],
                      ["data1", "data2", "data3"],
                      density=True,                      
                      ).plot()
```

## Further modifications of appearance

```python
fig = HistogramPlotter([data1, data2, data3],
                      ["data1", "data2", "data3"],
                      density=True,
                      figsize=(12,4),
                      ).plot()
```

```python
fig = HistogramPlotter([data1, data2, data3],
                      ["data1", "data2", "data3"],
                      density=True,
                      ).plot()

# Modify the size
fig.set_size_inches((12,4))

# Access the patches in the Axes object and modify alpha values for each histogram
fig_axes = fig.get_axes()
for ax in fig_axes:
    for patch in ax.patches:
        patch.set_alpha(0.2)

# Update the figure legend with the modified alpha values
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels);
```
